```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager
import matplotlib.ticker as ticker


colors = ['#ee1c25', '#303192', '#000000']

# Plot
plt.figure(figsize=(7, 5))

performance_10 = pd.read_csv('./results/ADR-10-10.csv')
performance_20 = pd.read_csv('./results/ADR-20-20.csv')
performance_30 = pd.read_csv('./results/ADR-30-30.csv')

radius = performance_10['radius']
error_10 = performance_10['error']
error_20 = performance_20['error']
error_30 = performance_30['error']

plt.plot(radius, error_10, label='(10, 10)', linestyle='--', color=colors[0], marker='s', markersize=3)
plt.plot(radius, error_20, label='(20, 20)', linestyle='-', color=colors[1], marker='D', markersize=3)
plt.plot(radius, error_30, label='(30, 30)', linestyle='-', color=colors[2], marker='^', markersize=3)

plt.xlabel('$\epsilon$')
plt.ylabel('$\delta$')



plt.ylim(0, 0.045)
ax = plt.gca()  
font = font_manager.FontProperties(family='Times New Roman', size=12)
ax.tick_params(axis='x', which='major', labelsize=12)
for label in ax.get_xticklabels():
    label.set_fontproperties(font)

ax.set_yticklabels(ax.get_yticks(), fontname='Times New Roman', fontsize=12)
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.3f'))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.legend(loc='best', prop=font)

# plt.savefig('./approximation.png', dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager
import matplotlib.ticker as ticker


performance_10 = pd.read_csv('./results/DRO-10-10.csv')
performance_20 = pd.read_csv('./results/DRO-20-20.csv')
performance_30 = pd.read_csv('./results/DRO-30-30.csv')

radius = performance_10['radius']
ratio_10 = performance_10['ratio']
ratio_20 = performance_20['ratio']
ratio_30 = performance_30['ratio']

performance_saa = pd.read_csv('./results/SAA.csv')
ratio_saa_10 = performance_saa['ratio'][0]
ratio_saa_20 = performance_saa['ratio'][1]
ratio_saa_30 = performance_saa['ratio'][2]

colors = ['#ee1c25', '#303192', '#000000']


plt.figure(figsize=(9, 6))


plt.plot(radius, ratio_10, label='(10, 10)-DRO', linestyle='-', color=colors[0], marker='s', markersize=3)
plt.plot(radius, ratio_20, label='(20, 20)-DRO', linestyle='-', color=colors[1], marker='s', markersize=3)
plt.plot(radius, ratio_30, label='(30, 30)-DRO', linestyle='-', color=colors[2], marker='s', markersize=3)

N = np.linspace(0, 1, 10)
plt.plot(N, [ratio_saa_10] * len(N), label='(10, 10)-SAA', linestyle='--', color=colors[0], marker='o', markersize=3)
plt.plot(N, [ratio_saa_20] * len(N), label='(20, 20)-SAA', linestyle='--', color=colors[1], marker='o', markersize=3)
plt.plot(N, [ratio_saa_30] * len(N), label='(30, 30)-SAA', linestyle='--', color=colors[2], marker='o', markersize=3)


plt.xlabel('$\epsilon$')
plt.ylabel('90th percentile ratios (%)', fontproperties=font)

plt.ylim(30, 135)

ax = plt.gca()  # Get the current axis
font = font_manager.FontProperties(family='Times New Roman', size=12)
ax.tick_params(axis='x', which='major', labelsize=12)
for label in ax.get_xticklabels():
    label.set_fontproperties(font)

ax.set_yticklabels(ax.get_yticks(), fontname='Times New Roman', fontsize=12)
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.0f'))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.legend(loc='best', prop=font)
# Show plot
# plt.savefig('./out-of-sample performance.png', dpi=300)
plt.show()
```

```python
data = pd.read_excel('./results/budget.xlsx', sheet_name='Sheet1')
fig, ax1 = plt.subplots(nrows=1, ncols=1, figsize=(12, 6))

ax1.bar([str(num) for num in data['budget'].values], data['build num'],color='#FFD06F', width=0.5, label='Number of built shelters')
ax1.set_xlabel('Budget')
ax1.set_ylabel('Number of built shelters')
ax1.set_yticks(range(0, max(data['build num']) + 5, 5))
# ax2 = ax1.twinx()

ax1.plot([str(num) for num in data['budget'].values], data['expanded num'], color='#E76254', linewidth=3, marker='*', label='Number of expanded shelters')
ax1.set_ylabel('Number of built (expanded) shelters')
ax1.set_yticks(ax1.get_yticks())
handles, labels = ax1.get_legend_handles_labels()

ax1.legend()
plt.gca()
# plt.show()
plt.grid()
plt.grid(False)
plt.savefig('../result/img/budget.png', dpi=300)
plt.show()
```
